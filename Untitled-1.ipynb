##### 1.3 Pre-Processing and making life easier

##### 1.3.1 Create support functions

One of the things that is required when looking at images in general, in order to really speed things up and reduce variability, there are standard techniques that I have learned that are used in the industry.
* Resize images
* Scale images/Normalize as it's more commonly referred to as (basically take the color images and make them between the 0-1 scale to reduce variable space)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tifffile as tiff
from skimage.transform import resize

# Function to read and display the image
def read_image(image_path):
    image = tiff.imread(image_path)
    plt.imshow(image)
    plt.colorbar()
    plt.title("Original Image")
    plt.show()
    return image

# Function to resize and display the image
def resize_image(image, output_shape=(64, 64)):
    resized_image = resize(image, output_shape, mode='constant', anti_aliasing=True)

    plt.imshow(resized_image)
    plt.title("Resized Image")
    plt.colorbar()
    plt.show()
    # Convert the image to float type
    resized_image = resized_image.astype(np.float32)
    return resized_image

# Function to normalize and display the image
def normalize_image(image):
    normalized_image = image / 255.0
    plt.imshow(normalized_image)
    plt.title("Normalized Image")
    plt.show()
    return normalized_image

# Combined preprocessing function
def preprocess_image(image_path, output_shape=(64, 64)):
    image = tiff.imread(image_path)
    resized_image = resize(image, output_shape, mode='constant', anti_aliasing=True)
    #normalized_image = resized_image / 255.0
    return resized_image#normalized_image

# Example usage
sample_image_path = 'train/50489f8b31b53756d56874d4c4e23b4b6b8636a8.tif'

# Read and visualize original image
sample_image = read_image(sample_image_path)

# Print the min and max values after loading
print(f'After loading: min={sample_image.min()}, max={sample_image.max()}')

# Resize and visualize resized image
resized_sample_image = resize_image(sample_image)

# Print the min and max values before normalization
print(f'Before normalization: min={resized_sample_image.min()}, max={resized_sample_image.max()}')
# Normalize and visualize normalized image
#normalized_sample_image = normalize_image(resized_sample_image)

# Full preprocessing and visualization
preprocessed_image = preprocess_image(sample_image_path)
plt.imshow(preprocessed_image)
plt.colorbar()
plt.title("Preprocessed Image")
plt.show()
```

#### 1.4 The first model

Typically we choose a simple and basic model for the first go, but we always try for best guess parameters. At this stage I am still a novice here and used class resources to put this together. 

Here we have a number of convolution layers followed by the pooling layers. The dropout and dense layers are for preventing overfitting and making predictions as per class notes.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

##### 1.4.1 Pre-process all, then split into train and validation sets

```python
from sklearn.model_selection import train_test_split

# Prepare paths and labels
image_paths = [f'train/{image_id}.tif' for image_id in train_labels['id']]
labels = train_labels['label'].values

# Train-test split
train_paths, val_paths, train_labels, val_labels = train_test_split(image_paths, labels, test_size=0.2, random_state=42)

# Data generators
def data_generator(image_paths, labels, batch_size=32):
    while True:
        batch_indices = np.random.choice(len(image_paths), batch_size)
        batch_images = [preprocess_image(image_paths[i]) for i in batch_indices]
        batch_labels = labels[batch_indices]
        yield np.array(batch_images), np.array(batch_labels)

train_gen = data_generator(train_paths, train_labels)
val_gen = data_generator(val_paths, val_labels)
```

##### 1.4.2 Let's run the model now

```python
# Training the model
history = model.fit(train_gen, 
                    steps_per_epoch=len(train_paths) // 32, 
                    validation_data=val_gen, 
                    validation_steps=len(val_paths) // 32, 
                    epochs=10)
```

```python
val_loss, val_accuracy = model.evaluate(val_gen, steps=len(val_paths) // 32)
print(f'Validation Accuracy: {val_accuracy}')
```

#### We can also plot the accuracy and losses

```python
# Plotting the training and validation accuracy and loss
def plot_training_history(history):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    epochs = range(len(acc))

    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.plot(epochs, acc, 'r', label='Training accuracy')
    plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
    plt.title('Training and validation accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(epochs, loss, 'r', label='Training loss')
    plt.plot(epochs, val_loss, 'b', label='Validation loss')
    plt.title('Training and validation loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    plt.show()

# Call the plotting function
plot_training_history(history)
```

## Trial Two - Hyper Parameter Tuning

#### Let's try things differently this time

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from keras_tuner import HyperModel, HyperParameters
from keras_tuner.tuners import RandomSearch
from sklearn.model_selection import train_test_split
import tifffile as tiff
from skimage.transform import resize

# Ensure TF uses the Metal backend
print(tf.config.list_physical_devices())

# Read and preprocess image function
def preprocess_image(image_path, output_shape=(64, 64)):
    image = tiff.imread(image_path)
    resized_image = resize(image, output_shape, mode='constant', anti_aliasing=True)
    #normalized_image = resized_image / 255.0
    return resized_image

# Data generator function
def data_generator(image_paths, labels, batch_size=32):
    while True:
        batch_indices = np.random.choice(len(image_paths), batch_size)
        batch_images = [preprocess_image(image_paths[i]) for i in batch_indices]
        batch_labels = labels[batch_indices]
        yield np.array(batch_images), np.array(batch_labels)
```

```python
# Load the labels
train_labels = pd.read_csv('train_labels.csv')

# Prepare paths and labels
image_paths = [f'train/{image_id}.tif' for image_id in train_labels['id']]
labels = train_labels['label'].values

# Use a smaller subset of the data for initial hyperparameter tuning
subset_size = 1000
image_paths = image_paths[:subset_size]
labels = labels[:subset_size]

# Train-test split
train_paths, val_paths, train_labels, val_labels = train_test_split(image_paths, labels, test_size=0.2, random_state=42)

# Data generators
train_gen = data_generator(train_paths, train_labels)
val_gen = data_generator(val_paths, val_labels)
```

#### What is different here is the model architecture, could we get away with a tweak to the model by lessening the steps? i.e. reducing the number of layers - the previous model took too long ? 

```python
#Define the model building function
def build_model(hp):
    model = Sequential()
    
    # Conv Layer 1
    model.add(Conv2D(
        filters=hp.Int('conv_1_filters', min_value=32, max_value=64, step=16),
        kernel_size=hp.Choice('conv_1_kernel_size', values=[3]),
        activation='relu',
        input_shape=(64, 64, 3)
    ))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    # Conv Layer 2
    model.add(Conv2D(
        filters=hp.Int('conv_2_filters', min_value=64, max_value=128, step=16),
        kernel_size=hp.Choice('conv_2_kernel_size', values=[3]),
        activation='relu'
    ))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    
    model.add(Flatten())
    
    # Fully Connected Layer
    model.add(Dense(
        units=hp.Int('dense_units', min_value=128, max_value=256, step=32),
        activation='relu'
    ))
    model.add(Dropout(hp.Float('dropout', min_value=0.2, max_value=0.4, step=0.1)))
    
    # Output Layer
    model.add(Dense(1, activation='sigmoid'))
    
    # Compile the model
    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    
    return model
```

#### Here we are going to run the parameter tuning. You will notice above I specified 1000 samples, as this gets really slow when running locally. 

```python
# Initialize the tuner
tuner = RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=10,
    executions_per_trial=1,
    directory='hyperparam_tuning',
    project_name='cancer_detection'
)
```

```python
# Run the search
tuner.search(train_gen, 
             steps_per_epoch=len(train_paths) // 32, 
             validation_data=val_gen, 
             validation_steps=len(val_paths) // 32, 
             epochs=5)
```

#### Model 2: Results
#### Ok with the model "tuned" we can evaluate the scores that it provides on the validation set

#### From before we can tell that this performed a bit more poorly than the first model. (89% vs 80% here - we did change the architecture and that could be the reason why). 

#### I suppose then this has taught me about the importance of the architectures and designing it properly. Will definitely pay more careful attention to it in the future.


```python
# Retrieve the best model
best_model = tuner.get_best_models(num_models=1)[0]

# Evaluate the best model on the validation data
val_loss, val_accuracy = best_model.evaluate(val_gen, steps=len(val_paths) // 32)
print(f'Best Model Validation Accuracy: {val_accuracy}')
```

```python
import os
test_image_paths = ['test/' + fname for fname in os.listdir('test')]

test_predictions = []
for image_path in test_image_paths:
    image_id = os.path.basename(image_path).split('.')[0]
    test_image = preprocess_image(image_path)
    prediction = best_model.predict(test_image.reshape(1, 64, 64, 3))
    predicted_label = int(prediction > 0.5)
    test_predictions.append([image_id, predicted_label])

submission_df = pd.DataFrame(test_predictions, columns=['id', 'label'])
submission_df.to_csv('submission.csv', index=False)
```
